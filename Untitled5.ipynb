```python
# Load the data from a CSV file
df = pd.read_csv(csv_file, dayfirst=True)

# Convert the 'date' column to a datetime object and set it as the index of the DataFrame
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# List of particle types
particle_types = [' pm10', ' o3', ' pm25']

print(df.columns)
```

```python
# Loop over particle types to handle missing data, sort data, and fill missing data
for particle_type in particle_types:
    # Handle missing values in the selected particle type column
    df[particle_type] = df[particle_type].apply(lambda x: float(x) if x.strip() != '' else np.nan)

    # Sort the DataFrame by date
    df.sort_index(inplace=True)
    
# Fill missing values with forward fill and then backward fill to handle NaNs at the start or end
for particle_type in particle_types:
    df[particle_type].fillna(method='ffill', inplace=True)
    df[particle_type].fillna(method='bfill', inplace=True)

    # Limit values to a maximum of 350
    df[particle_type] = df[particle_type].apply(lambda x: min(int(x), 350))

# Remove the columns we are not interested in
df = df[particle_types]
```

```python
# Reindex the DataFrame to ensure that it covers the entire date range
start_date = df.index.min()
end_date = df.index.max()
new_index = pd.date_range(start=start_date, end=end_date, freq=resample_frequency)
df = df.reindex(new_index)

# Fill missing values again after reindexing
for particle_type in particle_types:
    df[particle_type].fillna(method='ffill', inplace=True)
    df[particle_type].fillna(method='bfill', inplace=True)
```
