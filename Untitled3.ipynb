```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.impute import SimpleImputer
from sklearn import datasets
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
from datetime import datetime
import xgboost as xgb
```

```python
df = pd.read_csv('macedonia-air-quality1.csv', dayfirst=True)
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df[' pm10'] = df[' pm10'].apply(lambda x: float(x) if x.strip() != '' else np.nan)
```

```python
df.drop([' pm25', ' o3', ' no2', ' so2', ' co'], axis=1, inplace=True)
df.sort_index(inplace=True)
```

```python
# Handle missing values by mean imputation
imputer = SimpleImputer(strategy='mean')
df[' pm10'] = imputer.fit_transform(df[' pm10'].values.reshape(-1, 1))[:, 0]
# Limit unusual values
df[' pm10'] = df[' pm10'].apply(lambda x: min(int(x) if isinstance(x, str) and x.strip() != '' else int(x) if isinstance(x, float) else 0, 350))
```

```python
# Ensure that DataFrame covers entire date range
start_date = df.index.min()
end_date = df.index.max()
new_index = pd.date_range(start=start_date, end=end_date, freq='D')
df = df.reindex(new_index)

# Use mean imputation to fill missing values
imputer = SimpleImputer(strategy='mean')
df[' pm10'] = imputer.fit_transform(df[' pm10'].values.reshape(-1,1))[:,0]
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df[' pm10'])
plt.title('PM10 Particles Over Time')
plt.ylabel('PM10 Particle Level')
plt.grid(True)
plt.show()
```

```python
# Train-test split
train_size = int(len(df) * 0.7)
train, test = train_test_split(df, train_size=train_size, shuffle=False)
```

```python
def create_supervised_dataset(data, n_in):
    X, y = [], []
    for i in range(n_in, len(data)):
        X.append(data[i - n_in:i])
        y.append(data[i])
    return np.array(X), np.array(y)
```

```python
n_lag = 365  # Increase the lag to capture more history

X_train, y_train = create_supervised_dataset(train[' pm10'].values, n_lag)
X_test, y_test = create_supervised_dataset(test[' pm10'].values, n_lag)

model_xgboost = xgb.XGBRegressor(
    n_estimators=100,
    max_depth=3,
    learning_rate=0.1,
    objective='reg:squarederror'
)
model_xgboost.fit(X_train, y_train)
```

```python
# Predict the next 30 days
future_predictions = []
recent_data = df[' pm10'].values[-n_lag:].tolist()

for _ in range(365):
    # Predict the next value
    X_future = np.array(recent_data[-n_lag:]).reshape(1, -1)
    next_value = model_xgboost.predict(X_future)[0]
    future_predictions.append(next_value)

    # Update recent_data to include the next value
    recent_data.append(next_value)

# Convert future predictions to a pandas Series with a date range index
future_dates = pd.date_range(start=df.index.max() + pd.Timedelta(days=1), periods=365, freq='D')
future_series = pd.Series(future_predictions, index=future_dates)

# Plot the results
plt.figure(figsize=(14,7))
plt.plot(df.index[-365:], df[' pm10'][-365:], label='Historical Data')
plt.plot(future_series.index, future_series.values, label='Predicted Data')
plt.title('Air Quality Prediction for the Next 30 Days')
plt.xlabel('Date')
plt.ylabel('pm10')
plt.legend()
plt.grid(True)
plt.show()
```

```python
n_lag = 800  # Increase the lag to capture more history

X_train, y_train = create_supervised_dataset(train[' pm10'].values, n_lag)
X_test, y_test = create_supervised_dataset(test[' pm10'].values, n_lag)
```

```python
model_xgboost = xgb.XGBRegressor(
    n_estimators=100,
    max_depth=3,
    learning_rate=0.1,
    objective='reg:squarederror'
)
model_xgboost.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model_xgboost.predict(X_test)
```

```python
# Convert y_test and y_pred to pandas Series with appropriate index
y_test_series = pd.Series(y_test, index=test.index[n_lag:])
y_pred_series = pd.Series(y_pred, index=test.index[n_lag:])
```

```python
# Calculate metrics
mae_xgboost = mean_absolute_error(y_test, y_pred)
rmse_xgboost = np.sqrt(mean_squared_error(y_test, y_pred))
print('XGBoost MAE: %.3f' % mae_xgboost)
print('XGBoost RMSE: %.3f' % rmse_xgboost)
```

```python
# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(y_test_series.index, y_test_series, label='Actual')
plt.plot(y_pred_series.index, y_pred_series, color='red', label='Predicted')
plt.xlabel('Date')
plt.ylabel('pm10')
plt.title('PM10 Particle Predictions vs Actuals')
plt.legend()
plt.grid(True)
plt.show()
```
